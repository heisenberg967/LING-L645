```python
# imports

from datasets import load_dataset
import pandas as pd
import json
```

# Quizbowl Data

```python
# load qanta quizbowl dataset

quiz_bowl_data = load_dataset('qanta', 'mode=full,char_skip=25')

quizbowl_df = pd.DataFrame(quiz_bowl_data['buzztrain'])
```

```python
print(quizbowl_df.head())
```

```python
#print(len(quizbowl_df))

for item in set(quizbowl_df['category']):
  #print('category: ' + item)
  subset = quizbowl_df[quizbowl_df['category'] == item]
  #print('category length: ' + str(len(subset['text'])))
  for i in set(subset['subcategory']):
    #print('subcategory: ' + i)
    intent_data = subset[subset['subcategory'] == i]
    #print('subcategory length: ' + str(len(intent_data['text'])))
    #print('text example: ' + intent_data['text'].iloc[0])
    #print('\n \n')
```

```python
##### Test set

quizbowl_test_df = pd.DataFrame(quiz_bowl_data['buzztest'])

#print(len(quizbowl_test_df))

for item in set(quizbowl_test_df['category']):
  #print('category: ' + item)
  subset = quizbowl_test_df[quizbowl_test_df['category'] == item]
  #print('category length: ' + str(len(subset['text'])))
  for i in set(subset['subcategory']):
    #print('subcategory: ' + i)
    intent_data = subset[subset['subcategory'] == i]
    #print('subcategory length: ' + str(len(intent_data['text'])))
    #print('text example: ' + intent_data['text'].iloc[0])
    #print('\n \n')
```

```python
quizbowl_train_data = []

for item in set(quizbowl_df['category']):
  #print('category: ' + item)
  subset = quizbowl_df[quizbowl_df['category'] == item]
  #print('category length: ' + str(len(subset['text'])))
  for i in set(subset['subcategory']):
    intent_utterance_dict = {
        'intent': '',
        'utterances': []
      }
    intent_utterance_dict['intent'] = i
    #print('subcategory: ' + i)
    intent_data = subset[subset['subcategory'] == i]
    if len(intent_data['text']) > 50:
      for j in range(50):
        intent_utterance_dict['utterances'].append(intent_data['text'].iloc[j])
    else:
      for j in range(len(intent_data['text'])):
        intent_utterance_dict['utterances'].append(intent_data['text'].iloc[j])
    quizbowl_train_data.append(intent_utterance_dict)

# observe train data
print(quizbowl_train_data[:5])
```

```python
intent_utterance_pair = {
  }
for x in quizbowl_train_data:
  if x['intent'] not in intent_utterance_pair:
    intent_utterance_pair[x['intent']] = x['utterances']
  else:
    intent_utterance_pair[x['intent']].append(x['utterances'])
```

```python
for key in intent_utterance_pair:
  #print(key)
  utterances = intent_utterance_pair[key]
  for utterance in utterances:
    if len(utterance) > 490:
      #print(utterance)
      #print('\n')
      utterances.remove(utterance)
  intent_utterance_pair[key] = utterances
```

```python
quizbowl_test_data = []
quizbowl_test_df = pd.DataFrame(quiz_bowl_data['buzztest'])

for item in set(quizbowl_test_df['category']):
  #print('category: ' + item)
  subset = quizbowl_test_df[quizbowl_test_df['category'] == item]
  #print('category length: ' + str(len(subset['text'])))
  for i in set(subset['subcategory']):
    intent_utterance_dict = {
        'intent': '',
        'utterances': []
      }
    intent_utterance_dict['intent'] = i
    #print('subcategory: ' + i)
    intent_data = subset[subset['subcategory'] == i]
    if len(intent_data['text']) > 50:
      for j in range(50):
        intent_utterance_dict['utterances'].append(intent_data['text'].iloc[j])
    else:
      for j in range(len(intent_data['text'])):
        intent_utterance_dict['utterances'].append(intent_data['text'].iloc[j])
    quizbowl_test_data.append(intent_utterance_dict)

# observe test data
print(quizbowl_test_data[:5])
```

```python
# Export quizbowl data

with open('quizbowl_train_data.json', 'w') as outfile:
    json.dump(intent_utterance_pair, outfile)

with open('quizbowl_test_data.json', 'w') as outfile:
    json.dump(quizbowl_test_data, outfile)
```

# Wiki Data

```python
## Load our dataset on which models are to be evaluated

wiki_data = load_dataset('wiki_qa')
```

```python
# Observe train data sample
print(len(wiki_data['train']))
print(wiki_data['train'][0])
```

```python
wiki_train_df = pd.DataFrame(wiki_data['train'])
wiki_test_df = pd.DataFrame(wiki_data['test'])
```

```python
wiki_train_data = []

for item in set(wiki_train_df['document_title']):
  #print('category: ' + item)
  subset = wiki_train_df[wiki_train_df['document_title'] == item]
  intent_utterance_dict = {
        'intent': '',
        'utterances': []
      }
  intent_utterance_dict['intent'] = item
  for j in range(len(subset['answer'])):
    intent_utterance_dict['utterances'].append(subset['answer'].iloc[j])
  wiki_train_data.append(intent_utterance_dict)

# observe train data
print(wiki_train_data[:5])
```

```python
# doc title -> intent, answer -> utterances

wiki_test_data = []

for item in set(wiki_test_df['document_title']):
  #print('category: ' + item)
  subset = wiki_test_df[wiki_test_df['document_title'] == item]
  intent_utterance_dict = {
        'intent': '',
        'utterances': []
      }
  intent_utterance_dict['intent'] = item
  for j in range(len(subset['answer'])):
    intent_utterance_dict['utterances'].append(subset['answer'].iloc[j])
  wiki_test_data.append(intent_utterance_dict)

# observe test data
print(wiki_test_data[:5])
```

```python
# Export wiki data

with open('wiki_train_data.json', 'w') as outfile:
    json.dump(wiki_train_data, outfile)

with open('wiki_test_data.json', 'w') as outfile:
    json.dump(wiki_test_data, outfile)
```
